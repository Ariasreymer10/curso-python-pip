Functions



Python functions are reusable blocks of code that allow you to organize and structure your code in a modular and efficient way. Functions can receive input data, perform operations and return results.

Defining a Function
To define a function in Python, the keyword "def" is used followed by the function name and parentheses that may contain parameters. The code block inside the function must be indented.


````python
#Structure of a Functions in Python

def function_name(parameter1, parameter2):
    #Code block
    result = parameter1 + parameter2
    return result

#Calling the function

print(function_name(2, 3))
````

***Parameter and Arguments*** 



Parameters: These are the variables that are defined in the function declaration and act as containers for the values that are passed to the function. 

Arguments: These are the values that are passed to the function when it is called.

***Argumentos Posicionales***

Los argumentos posicionales en una funcin de Python son aquellos que se pasan a la funcin en el orden en el que estn definidos en la declaracin de la funcin. Estos argumentos se asignan a los parmetros de la funcin en el mismo orden.

Por ejemplo, considera la siguiente funcin:

```python
def sumar(a, b):
    return a + b
```

En esta funcin, `a` y `b` son los parmetros que esperan recibir dos argumentos. Si llamamos a la funcin de la siguiente manera:

```python
resultado = sumar(3, 5)
```

El valor `3` se asignar al parmetro `a` y el valor `5` se asignar al parmetro `b`. La funcin devolver el resultado de sumar `3` y `5`, que es `8`.

Es importante tener en cuenta que el orden de los argumentos posicionales es crucial. Si cambiamos el orden de los argumentos al llamar a la funcin, obtendremos un resultado diferente:

```python
resultado = sumar(5, 3)
```

En este caso, el valor `5` se asignar al parmetro `a` y el valor `3` se asignar al parmetro `b`. La funcin devolver el resultado de sumar `5` y `3`, que es `8`.

En resumen, los argumentos posicionales en una funcin de Python se pasan en el orden en el que estn definidos en la declaracin de la funcin y se asignan a los parmetros correspondientes en ese mismo orden.

****Argumentos Nombrados o Argumentos por Nombre****

Los Argumentos se Asignan a los parmetros por nombre, lo que permite cambiar el orden de estos.

**Valores por Defecto** 

Se puede definir valores por defecto para los parmetros, si no se proporciona un argumento para un parmetro con el valor por defecto. Se puede Establecer valores por defecto en caso no se ingrese un valor para poder dearrollar el funcionamiento de la funcin 

````python
chiefs = ['John', 'Jane', 'Doe', 'Doe']
positions = ['CEO', 'CTO', 'CFO', 'COO']

directory = dict(zip(positions, chiefs))  # Diccionario ms intuitivo

def report_problem(position, msg="Error general"):
    """
    Funcin para reportar un problema al jefe correspondiente.

    Args:
        position (str): Posicin del jefe (CEO, CTO, CFO, COO).
        msg (str, optional): Descripcin del problema. Por defecto es "Error general".
    """

    if position in directory:  # Verificar si la posicin existe
        chief = directory[position]
        print(f"Problema: '{msg}' reportado a {chief} ({position})")
    else:
        print(f"Error: Posicin '{position}' no encontrada.")

# Ejemplo de uso:
report_problem("CEO", "Servidor cado")
report_problem("CTO", "Fallo en el sistema de autenticacin")
report_problem("COO", "Retraso en la produccin")
report_problem("Gerente de Marketing")  # Prueba de error
````

````python
#The Global Keyword 
#The Global Keyword is used to modify a global variable with function 

d = 40

def modify_global_function():
    global d
    d = 50

print(d)  # Output: 40
modify_global_function()
print(d)  # Output: 50
````

````python
'''English Translation:

The Nonlocal Keyword
The nonlocal keyword is used to modify a variable in the enclosing scope (neither local nor global).'''

def enclosing_nonlocal_function():
    e = 60
    def inner_nonlocal_function():
        nonlocal e
        e = 70
    inner_nonlocal_function()
    print(e)

enclosing_nonlocal_function()  # Output: 70

'''
Here, e is redefined within inner_nonlocal_function using nonlocal, thus affecting its value in the enclosing function.
'''
````

# <span style="color: red;">LAMBDA FUNCTIONS</span>

An <span style="font-weight: bold;">Anonymous function</span> is a function that is defined without a name. In Python, anonymous functions are created using the <span style="font-weight: bold;">lambda</span> keyword. Unlike regular functions defined with <span style="font-weight: bold;">"def"</span>, lambda functions are limited to a single expression and are designed to be simple and temporary in use.

<span style="font-weight: bold;">Python lambdas are only a shorthand notation if youre too lazy to define a function.</span>


```python
lambda argument : expression
```
<span style="font-weight: bold;">lambda</span> is the keyword that initiates the definition of the anonymous function.

<span style="font-weight: bold;">arguments</span> is a list of parameters that the function takes.

<span style="font-weight: bold;">expression</span> is an expression that is evaluated and returned as the result of the function.
